```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/health_lifestyle_data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(data=df, x='steps', y='calories',hue='gender')
plt.savefig('scatterplot.png')
```

```python
sns.lineplot(data=df, x='date', y='sleep_hours')
plt.savefig('lineplot.png')
```

```python
sns.countplot(data=df, x='gender',palette='pastel')
plt.savefig('countplot.png')
```

```python
sns.barplot(data=df, x='age_group', y='sleep_hours', palette='Set2')
plt.savefig('barplot.png')
```

```python
sns.histplot(df['heart_rate'])
plt.savefig('histplot.png')
```

```python
sns.kdeplot(df['calories'],fill=True)
plt.savefig('kdeplot.png')
```

```python
sns.boxplot(data=df, x='gender', y='calories', hue='gender')
plt.savefig('boxplot.png')
```

```python
sns.violinplot(data=df, x='gender', y='calories',palette='bright')
plt.savefig('violinplot.png')
```

```python
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.savefig('heatplot.png')
```

```python
sns.pairplot(numeric_df)
plt.savefig('pairplot.png')
```
