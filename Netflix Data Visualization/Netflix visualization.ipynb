```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('/content/Netflix.csv')
```

```python
#cleaning
df=df.dropna(subset=['type','release_year','country','duration'])
```

```python
#bar graph
type_counts=df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index,type_counts.values, color=['red','orange'])
plt.title('Number ofTVs and Movies comparision')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('bar.png')
plt.show()
```

```python
#pie chart
rating_counts=df['rating'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_counts.values, labels=rating_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('TV Show ratings')
plt.tight_layout()
plt.savefig('pie.png')
plt.show()
```

```python
# histogram
movie_df = df[df['type']== 'Movie'].copy()
movie_df['duration'] = movie_df['duration'].str.replace(' min', '').astype(int)

plt.figure(figsize=(8,6))
plt.hist(movie_df['duration'], bins=30, color='purple', edgecolor='black')
plt.title('Duration of movies')
plt.xlabel('Duration(in min)')
plt.ylabel('Movies')
plt.tight_layout()
plt.savefig('hist.png')
plt.show()
```

```python
# scatter plot
release_year_counts = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(6,4))
plt.scatter(release_year_counts.index, release_year_counts.values, color='red')
plt.title('Release year vs shows')
plt.xlabel('release year')
plt.ylabel('shows')
plt.tight_layout()
plt.savefig('scatter.png')
plt.show()
```

```python
# horizontal bar
#bar graph
country_counts=df['country'].value_counts().head(10)
plt.figure(figsize=(6,4))
plt.barh(country_counts.index,country_counts.values, color='teal')
plt.title('Top 10 country relase shows')
plt.xlabel('Count')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig('barh.png')
plt.show()
```

```python
# subplot
content_by_year = df.groupby(['release_year','type']).size().unstack().fillna(0)
fig, axs = plt.subplots(1,2, figsize=(12,6))

axs[0].plot(content_by_year.index, content_by_year['Movie'], color='blue')
axs[0].set_title('Movies Released By Year')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Count')

axs[1].plot(content_by_year.index, content_by_year['TV Show'], color='grey')
axs[1].set_title('TV Shows Released By Year')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Count')

plt.tight_layout()
plt.savefig('subplot.png')
plt.show()
```
